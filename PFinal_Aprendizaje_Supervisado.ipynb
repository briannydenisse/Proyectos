**Cree un conjunto de entrenamiento y otro de test a partir del conjunto inicial de datos donde el conjunto de entrenamiento contenga el 60% de las observaciones.**

```python
#Muestreo
from sklearn.model_selection import train_test_split
X = data.loc[:, data.columns != "Target"]
y = data.loc[:, data.columns == "Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40, random_state = 0)
```

#Fase 3: Ajuste del algoritmo de Naive Bayes.

**A partir del conjunto de entrenamiento ajuste un modelo a partir del algoritmo de Naive Bayes.**

```python
#Ajuste del modelo
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

modelNB = gnb.fit(X_train, y_train.values.ravel())
y_pred_train = modelNB.predict_proba(X_train)
y_pred_test = modelNB.predict_proba(X_test)
```

**Obtenga la curva ROC y el √°rea bajo la curva para los conjuntos de entrenamiento y test.**

```python
#Importar librerias
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, auc
```

```python
#Curva ROC y AUC

#Training
fpr_train, tpr_train, _ = roc_curve(y_train, y_pred_train[:, 1])
roc_auc_train = auc(fpr_train, tpr_train)

#Test
fpr_test, tpr_test, _ = roc_curve(y_test, y_pred_test[:, 1])
roc_auc_test = auc(fpr_test, tpr_test)
```

```python
#Grafica
plt.figure()
lw = 2

plt.plot(fpr_train,tpr_train,color="darkgreen",lw=lw,label="ROC curve (area = %0.2f)" % roc_auc_train,)
plt.plot(fpr_test,tpr_test,color="darkred",lw=lw,label="ROC curve (area = %0.2f)" % roc_auc_test,)

plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC")
plt.legend(loc="lower right")
plt.show()
```

# **Adicional:** B√∫squeda de la mejor combinaci√≥n de hiperpar√°metros en el algoritmo de SVM utilizando la funci√≥n Grid Search de Scikit Learn.

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Hiperparametros
param_grid = [
    {"kernel": ["rbf"], "gamma": [1e-3, 1e-4], "C": [0.1,1,10]},
    {"kernel": ["linear"], "C": [0.1,1,10]},
    {"kernel": ["poly"], "C": [0.1,1,10], "degree": [2,3]},
]

grid = GridSearchCV(
        estimator  = SVC(),
        param_grid = param_grid,
        scoring    = 'roc_auc',
        n_jobs     = -1,
        cv         = 3,
        verbose    = 0,
        return_train_score = True
      )

grid.fit(X = X_train_svm, y = y_train_svm)
```

```python
# Resultados del grid

resultados = pd.DataFrame(grid.cv_results_)
resultados.filter(regex = '(param.*|mean_t|std_t)')\
    .drop(columns = 'params')\
    .sort_values('mean_test_score', ascending = False) \
    .head(10)
```

```python
# mejores parametros
print(grid.best_params_)

# como luce el modelo
print(grid.best_estimator_)
```

# Tratamiento de Outliers

```python
#Eliminacion de Outliers
Q1 = data['ALTV'].quantile(0.25)
Q3 = data['ALTV'].quantile(0.75)
IQR = Q3 - Q1
df_filtered = data[(data['ALTV'] >= Q1 - 1.5 * IQR) & (data['ALTV'] <= Q3 + 1.5 * IQR)]
```

```python
#Imputaci√≥n de Outliers con valores mas tipicos
Q1 = data['UC'].quantile(0.25)
Q3 = data['UC'].quantile(0.75)
IQR = Q3 - Q1


median_value = data['UC'].median()

df_filtered.loc[:, 'UC'] = np.where((df_filtered['UC'] < Q1 - 1.5 * IQR) | (df_filtered['UC'] > Q3 + 1.5 * IQR), median_value, df_filtered['UC'])
```

```python
#Transformaci√≥n de Variables
df_filtered['AC'] = np.log1p(df_filtered['AC'])
```

```python
#Muestreo con nuevo df
from sklearn.model_selection import train_test_split
X_nuevo = df_filtered.loc[:, df_filtered.columns != "Target"]
y_nuevo = df_filtered.loc[:, df_filtered.columns == "Target"]

X_train_nuevo, X_test_nuevo, y_train_nuevo, y_test_nuevo = train_test_split(X_nuevo, y_nuevo, test_size = 0.40, random_state = 0)
```

```python
#busqueda mejores parametros para nuevo df
# Hiperparametros
param_grid = [
    {"kernel": ["rbf"], "gamma": [1e-3, 1e-4], "C": [0.1,1,10]},
    {"kernel": ["linear"], "C": [0.1,1,10]},
    {"kernel": ["poly"], "C": [0.1,1,10], "degree": [2,3]},
]

grid = GridSearchCV(
        estimator  = SVC(),
        param_grid = param_grid,
        scoring    = 'roc_auc',
        n_jobs     = -1,
        cv         = 3,
        verbose    = 0,
        return_train_score = True
      )

grid.fit(X = X_train_nuevo, y = y_train_nuevo)
```

```python
# mejores parametros
print(grid.best_params_)

# como luce el modelo
print(grid.best_estimator_)
```
