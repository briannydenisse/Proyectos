```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('incident.csv')
```

```python
df.head()
```

```python
#el n√∫mero de incidentes por estado.

status_counts = df['status'].value_counts()

plt.figure(figsize=(10, 6))
status_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('N√∫mero de Incidentes por Estado')
plt.xlabel('Estado')
plt.ylabel('N√∫mero de Incidentes')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show
```

```python
# Contar el n√∫mero de incidentes por tipo
incident_type_counts = df['incident_type'].value_counts()


plt.figure(figsize=(10, 6))
incident_type_counts.plot(kind='bar', color=['red', 'green', 'blue'])
plt.title('N√∫mero de Incidentes por Tipo')
plt.xlabel('Tipo de Incidente')
plt.ylabel('N√∫mero de Incidentes')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
#el promedio de tiempo de respuesta de incidentes por tipo

average_response_time = df.groupby('incident_type')['response_time_minutes'].mean()

# Crear la gr√°fica
plt.figure(figsize=(10, 6))
average_response_time.plot(kind='bar', color=['red', 'green', 'blue'])
plt.title('Promedio de Tiempo de Respuesta por Tipo de Incidente')
plt.xlabel('Tipo de Incidente')
plt.ylabel('Promedio de Tiempo de Respuesta (minutos)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
#el n√∫mero de incidentes por equipo de respuesta
response_team_counts = df['response_team_id'].value_counts()


plt.figure(figsize=(12, 6))
response_team_counts.plot(kind='bar', color='purple')
plt.title('N√∫mero de Incidentes por Equipo de Respuesta')
plt.xlabel('Equipo de Respuesta')
plt.ylabel('N√∫mero de Incidentes')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
# El promedio de tiempo de respuesta por equipo de respuesta
average_response_time_by_team = df.groupby('response_team_id')['response_time_minutes'].mean()

plt.figure(figsize=(12, 6))
average_response_time_by_team.plot(kind='bar', color='cyan')
plt.title('Promedio de Tiempo de Respuesta por Equipo de Respuesta')
plt.xlabel('Equipo de Respuesta')
plt.ylabel('Promedio de Tiempo de Respuesta (minutos)')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```
